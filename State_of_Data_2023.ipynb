```python
df_salarios = new_df[(['Estado','Faixa Salarial','Nivel de ensino'])]
```

```python



# Remova as linhas onde o valor da coluna 'Estados' é 0.0
df_salarios = df_salarios[df_salarios['Estado'] != 0.0]

# Exiba o dataframe atualizado
print(df_salarios)
```

```python
df_salarios['Faixa Salarial'].value_counts()
```

```python


# Mapeie as faixas salariais existentes para novas categorias
faixas_remodeladas = {
    'de R$ 8.001/mês a R$ 12.000/mês': '8K-12K/ mês',
    'de R$ 4.001/mês a R$ 6.000/mês': '4K-6K/ mês',
    'de R$ 6.001/mês a R$ 8.000/mês': '6K-8K/mês',
    'de R$ 12.001/mês a R$ 16.000/mês': '12K-16K /mês',
    'de R$ 3.001/mês a R$ 4.000/mês': '3K-4K /mês',
    'de R$ 2.001/mês a R$ 3.000/mês': '2K-3K /mês',
    'de R$ 1.001/mês a R$ 2.000/mês': '1K-2K /mês',
    'de R$ 16.001/mês a R$ 20.000/mês': '16K-20K /mês',
    'de R$ 20.001/mês a R$ 25.000/mês': '20K-25K/ mês',
    'de R$ 25.001/mês a R$ 30.000/mês': '25K-30K /mês',
    'Acima de R$ 40.001/mês': '>40K /mês',
    'de R$ 30.001/mês a R$ 40.000/mês': '30K-40K /mês',
    'Menos de R$ 1.000/mês': '<1K/ mês'
}

# Remodele a coluna "faixa salarial" usando o mapeamento
df_salarios['Faixa Salarial'] = df_salarios['Faixa Salarial'].map(faixas_remodeladas)

# Exiba o dataframe atualizado
print(df_salarios)
```

```python


# Remover caracteres não numéricos da coluna 'Faixa Salarial' e converter para tipo numérico
df_salarios['Faixa Salarial'] = df_salarios['Faixa Salarial'].str.replace('[^\d.]', '', regex=True).astype(float)

# Percorrer o DataFrame e encontrar os estados que pagam mais
estados_pagam_mais = []
max_salario_por_estado = df_salarios.groupby('Estado')['Faixa Salarial'].max()
for estado, salario_maximo in max_salario_por_estado.items():
    estados_pagam_mais.append(estado)

# Imprimir os estados que pagam mais
print("Estados que pagam os salários mais altos:")
for estado in estados_pagam_mais:
    print(estado)
```

```python


# Criação do DataFrame de exemplo
data =[["('P1_a_1 ', 'Faixa idade')"]]
df = pd.DataFrame(data)




# Função para dividir os valores e selecionar apenas o primeiro número
def split_and_select_first(value):
    if isinstance(value, str):
        parts = value.split('-')
        if len(parts) > 0:
            return parts[0]
    return value

# Aplicar a função em todas as colunas do DataFrame
df = df.applymap(split_and_select_first)

# Converter as colunas para tipo numérico
df = df.apply(pd.to_numeric, errors='ignore')

# Calcular a média das colunas numéricas
media = df.mean(numeric_only=True)

# Exibir a média de cada coluna
print(media)

```
